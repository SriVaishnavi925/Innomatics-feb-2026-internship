```python
# Problem 1: Employee Performance Bonus Eligibility
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

highest_score = max(employees.values())
# Identify all employees with the highest score
top_performers = []
for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

```python
# Problem 2: Search Query Keyword Analysis

from collections import Counter

query = "Buy mobile phone buy phone online"

words = query.lower().split()
keyword_count = Counter(words)

# Filter for keywords that appear more than once
result = {k: v for k, v in keyword_count.items() if v > 1}
print(result)
```

```python
# Problem 3: Sensor Data Validation
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for hour, value in enumerate(sensor_readings):
    # Check if the sensor reading is an even number
    if value % 2 == 0:
        valid_readings.append((hour, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Problem 4: Email Domain Usage Analysis

from collections import Counter

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Extract domains using a list comprehension
domains = [email.split("@")[1] for email in emails]

# Count domain occurrences using Counter
domain_count = Counter(domains)

total = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# Problem 5: Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

average_sales = sum(sales) / len(sales)
threshold = average_sales * 1.3

spikes_found = False
for day, value in enumerate(sales, start=1):
    # Check if the current day's sales exceed the threshold
    if value > threshold:
        print(f"Day {day}: {value}")
        spikes_found = True

if not spikes_found:
    print("No sales spikes detected above the threshold.")
```

```python
# Problem 6: Duplicate User ID Detection
from collections import Counter

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = Counter(user_ids)

for uid, count in id_count.items():
    if count > 1:
        print(f"{uid} → {count} times")
```
