```python
# Task 1: User Login Check

username = "admin"
password = "1234"

entered_user = input("Enter username: ")
entered_pass = input("Enter password: ")
# Check if both credentials match
if entered_user == username and entered_pass == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# Task 2: Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0
# Count pass and fail students
for m in marks:
    if m >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Passed Students:", pass_count)
print("Total Failed Students:", fail_count)
```

```python
# Task 3: Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []
# Clean each name by removing spaces and converting to lowercase
for name in names:
    cleaned_names.append(name.strip().lower())

print("Cleaned Names List:", cleaned_names)
```

```python
# Task 4: Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    print("Message:", msg)
    print("Length:", len(msg))
    # Flag long messages
    if len(msg) > 10:
        print("Flag: Long message")

    print()
```

```python
# Task 5: Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0
# Count how many times "ERROR" appears
for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total ERROR entries:", error_count)
```
