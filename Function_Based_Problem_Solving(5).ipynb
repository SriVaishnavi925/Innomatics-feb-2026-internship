```python
# Problem 1: Social Media – Post Engagement Analyzer
def analyze_post_engagement(engagement_data):
    total_likes = 0
    for like_count in engagement_data:
        total_likes += like_count

    # Determine the status based on the 1000 likes
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")

#For example
performance_stats = [450, 200, 100, 350]
analyze_post_engagement(performance_stats)
```

```python
# Problem 2: Healthcare – Medicine Stock Alert System
def monitor_medicine_stock(stock_level):
    if stock_level < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print(f"Medicine Stock: {stock_level}")
    print(f"Status: {status}")

# for example
monitor_medicine_stock(6)
```

```python
# Problem 3: Agriculture – Rainfall Adequacy Checker
def check_rainfall_adequacy(rainfall_data, required_level):
    total_rainfall = 0
    count = 0
    for amount in rainfall_data:
        total_rainfall += amount
        count += 1

    average_rainfall = total_rainfall / count

    # Determine status
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print(f"Average Rainfall: {int(average_rainfall)}")
    print(f"Rainfall Status: {status}")

#  For Example
check_rainfall_adequacy([82, 68, 51, 74], 70)
```

```python
# Problem 4: Social Media – Duplicate Account Detection
def detect_duplicates(usernames):
    unique_usernames = set(usernames)
    if len(usernames) != len(unique_usernames):
        status = "Yes"
    else:
        status = "No"
    print(f"Duplicate Accounts Found: {status}")

# For Example
detect_duplicates(["user1", "user2", "user2", "user3"])
```

```python
# Problem 5: Healthcare – Appointment Eligibility Checker
def check_appointment_eligibility(patient_age):
    if patient_age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"
    print(f"Patient Age: {patient_age}")
    print(f"Eligibility Status: {status}")

#  For Example
check_appointment_eligibility(24)
```

```python
# Problem 6: Agriculture – Premium Crop Price Filter
def premium_crops(crop_prices):
    threshold = 2000
    premium_list = [price for price in crop_prices if price > threshold]
    print(f"Premium Crops: {premium_list}")

# For Example
market_prices = [1800, 4500, 1600, 2200, 1000]
premium_crops(market_prices)
```

```python
# Problem 7: System Monitoring – Application Health Checker
def check_app_health(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print(f"Error Count: {error_count}")
    print(f"System Status: {status}")

#  For Example
check_app_health(8)
```

```python
# Problem 8: Banking – Daily Transaction Limit Checker
def validate_transaction(transaction_amount):
    DAILY_LIMIT = 50000
    if transaction_amount <= DAILY_LIMIT:
        status = "Approved"
    else:
        status = "Rejected"

    print(f"Transaction Amount: {transaction_amount}")
    print(f"Transaction Status: {status}")

# For Example
validate_transaction(45000)
```

```python
# Problem 9: E-Learning – Student Attendance Eligibility System
def check_exam_eligibility(attendance_list):
    total_sessions = len(attendance_list)
    attended_sessions = 0

    # (1 = Present, 0 = Absent)
    for status in attendance_list:
        if status == 1:
            attended_sessions += 1

    attendance_percentage = (attended_sessions / total_sessions) * 100
    if attendance_percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"
    print(f"Attendance Percentage: {attendance_percentage:.1f}")
    print(f"Exam Eligibility: {eligibility}")

# For Example (1 for present, 0 for absent)
student_attendance = [1, 1, 0, 1, 1]
check_exam_eligibility(student_attendance)
```

```python
# Problem 10: Smart Electricity Bill Analyzer
def analyze_electricity_bill(units):
    bill = 0
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    # Classification based on total bill amount
    if bill < 500:
        usage_status = "Low Usage"
    elif 500 <= bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"

    return bill, usage_status

#  For Example
units_consumed = 150
total_bill, status = analyze_electricity_bill(units_consumed)

print(f"Total Bill: ₹{total_bill}")
print(f"Usage Status: {status}")
```
