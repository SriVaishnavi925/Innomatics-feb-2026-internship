```python
#problem Statement 1:Smart Parking Lot Management System
def smart_parking_system(capacity, logs):
    current_vehicles = 0
    peak_usage = 0
    for action in logs:
        if action == "IN":
            current_vehicles += 1
            if current_vehicles > peak_usage:
                peak_usage = current_vehicles
        elif action == "OUT":
            current_vehicles -= 1
    if current_vehicles >= capacity:
        status = "Alert: Capacity Exceeded"
    else:
        status = "Available"
    print(f"Currently Parked Vehicles: {current_vehicles}")
    print(f"Peak Parking Usage: {peak_usage}")
    print(f"Parking Status: {status}")

# Sample Input
parking_capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
smart_parking_system(parking_capacity, vehicle_logs)
```

```python
# Problem Statement 2:Online Food Delivery Time Estimator
def estimate_delivery_time(distance, traffic, weather):

    base_time = distance * 5

    traffic_delay = 0
    weather_delay = 0

    if traffic.lower() == "low":
        traffic_delay = 0.10
    elif traffic.lower() == "medium":
        traffic_delay = 0.20
    elif traffic.lower() == "high":
        traffic_delay = 0.30

    if weather.lower() == "rainy":
        weather_delay = 0.15
    elif weather.lower() == "storm":
        weather_delay = 0.30

    total_time = base_time + (base_time * traffic_delay) + (base_time * weather_delay)

    return total_time

distance = float(input("Distance (km): "))
traffic = input("Traffic Level (Low/Medium/High): ")
weather = input("Weather (Rainy/Storm): ")

eta = estimate_delivery_time(distance, traffic, weather)

print("Estimated Delivery Time:", int(eta), "minutes")
```

```python
# Problem Statement 3:Movie Theatre Seat Occupancy Analyzer
def analyze_seats(total_seats, booked_seats):

    booked = len(booked_seats)
    occupancy = (booked / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
        suggestion = "Open additional show"
    elif occupancy >= 75:
        status = "Almost Full"
        suggestion = "Consider opening additional show"
    else:
        status = "Seats Available"
        suggestion = "No need for extra show"

    print("Occupancy:", int(occupancy), "%")
    print("Show Status:", status)
    print("Suggestion:", suggestion)


total = int(input("Total Seats: "))
n = int(input("Number of booked seats: "))

booked_list = []

for i in range(n):
    booked_list.append(1)

analyze_seats(total, booked_list)
```

```python
#Problem Statement 4:Cloud Server Load Classification System
def classify_server_load(cpu_readings):

    total = 0

    for i in cpu_readings:
        total += i

    average = total / len(cpu_readings)

    if average < 50:
        status = "Normal"
    elif average < 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", int(average), "%")
    print("Server Status:", status)


cpu_readings = [45, 60, 70, 85, 90]

classify_server_load(cpu_readings)
```

```python
 #Problem Statement 5:Smart Classroom Resource Usage Monitor
def monitor_resources(resource_usage):

    overused = []

    for resource in resource_usage:
        if resource_usage[resource] > 8:
            overused.append(resource)

    if len(overused) > 0:
        alert = "Yes"
    else:
        alert = "No"

    if len(overused) > 0:
        print("Overused Resources:", ", ".join(overused))
    else:
        print("Overused Resources: None")

    print("Energy Alert:", alert)


resources = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resources(resources)
```

```python
#Problem Statement 6: Online Event Registration Capacity Controller
def manage_registrations(capacity, total_registrations):

    confirmed = 0
    waitlisted = 0

    for i in range(total_registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    if confirmed == capacity:
        status = "Closed"
    else:
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


event_capacity = 200
registrations = 210

manage_registrations(event_capacity, registrations)
```
