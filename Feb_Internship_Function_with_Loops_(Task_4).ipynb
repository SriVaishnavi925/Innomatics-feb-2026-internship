```python
# Problem 1: Mobile Recharge Validation System
def validate_recharge(amount):
    plans = [199, 299, 399, 599]

    if amount < 50:
        return False

    if amount not in plans:
        return False

    return True


while True:
    amt = int(input("Enter recharge amount: "))

    if validate_recharge(amt):
        print("Recharge Successful")
        break
    else:
        print("Invalid recharge amount. Try again.")
```

```python
# Problem 2: Inventory Reorder Alert System
def check_inventory(stock):

    for product in stock:
        if stock[product] < 15:
            print(product, ": Reorder Alert")
        else:
            print(product, ": Stock OK")


# Example dictionary
products = {
    "Rice": 10,
    "Wheat": 25,
    "Sugar": 12,
    "Oil": 30
}

check_inventory(products)
```

```python
# Problem 3: Student Result Processing System
def process_result(marks):
    if len(marks) == 0:
        return "No marks provided"
    total = 0
    # Calculate total using loop
    for mark in marks:
        total += mark
    average = total / len(marks)
    print("Average Marks:", average)
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

#example
student_marks = [60, 45, 70, 55, 40]
result = process_result(student_marks)
print("Result:", result)
```

```python
# Problem 4: Cab Fare Estimator with Retry Option
def calculate_fare(km, peak_hour):

    base_fare = 50
    per_km = 12
    fare = base_fare + (km * per_km)
    if peak_hour == True:
        fare = fare + (fare * 0.25)
    return fare

while True:
    km = float(input("Enter distance in km: "))
    peak = input("Peak hour? (yes/no): ").lower()
    if peak == "yes":
        peak_hour = True
    else:
        peak_hour = False
    total_fare = calculate_fare(km, peak_hour)
    print("Total Fare: ₹", total_fare)
    choice = input("Calculate again? (yes/no): ").lower()

    if choice != "yes":
        break
```

```python
# Problem 5: Employee Attendance Eligibility Checker
def check_attendance(attendance):
    total_days = len(attendance)
    present_days = 0
    # Count present days using loop
    for day in attendance:
        if day == "P":
            present_days += 1
    percentage = (present_days / total_days) * 100
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Example
attendance_list = ["P", "A", "P", "P", "P", "A", "P", "P"]
print(check_attendance(attendance_list))
```

```python
# Problem 6: Password Strength Checker
def password_strength(password):
    special_chars = "@#$"
    has_digit = False
    has_special = False
    if len(password) < 8:
        return "Weak Password"
    # Loop through each character
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

pwd = input("Enter password: ")
print(password_strength(pwd))
```
